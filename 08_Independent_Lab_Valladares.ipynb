## Rename columns

```python
(gh_turnover
 .rename({'domain':'project_domain',
         'language':'project_language',
         'windows':'project_age',
         'window_idx':'quarter',
         'num_team':'team_size',
         'num_commits':'project_commits',
         'blau_gender':'gender_blau',
         'Gini_gh_ten':'gh_tenure_gini',
         'Core1':'core_dev',
         'commits':'user_commits',
         'propCommits':'commit_proportion',
         'github_tenure':'gh_tenure',
         'leavesNextQ':'leaves_next_q'},  
        axis = 1,
        inplace = True) # Rename columns
)
gh_turnover.info()
```

## Data Wrangling

```python
# Calculate the mean of `user_commits`
(gh_turnover
 .filter(["user_commits"])
 .agg(["mean"])
)
```

```python
# Calculate the median of `user_commits`
(gh_turnover
 .filter(["user_commits"])
 .agg(["median"])
)
```

```python
# Select all columns that start with a *p* (i.e., `(^p)`) **or** contain an *g* (i.e., `(g)`). Save it as a new DataFrame named `turnover_new`. Output the columns in your notebook.
turnover_new = (gh_turnover
          .filter(regex = "(^p)|(g)")  # select desired columns
         )

turnover_new.info()
```

```python
# Using your newly created DataFrame `p_or_g`, select rows in which `project_commits` is greater than 9,000.
(turnover_new
 .query("project_commits > 9000") # keep rows based on value the criteria 
 )
```

```python
# Using pandas piping notation, perform the previous two operations together and save it as a new DataFrame `turnover_newer`. 
# This means you should select columns that start with a *p* or contain an *g* and select rows in which `project_commits` is greater than 9,000
turnover_newer = (gh_turnover
                  .filter(regex = "(^p)|(g)")  # select desired columns
                  .query("project_commits > 9000") # keep rows based on value the criteria 
                 )
turnover_newer
```

```python
# Using pandas piping notation, select all columns that end with the letter *e* (i.e., `e$`) or contain the letter *q* (i.e., `s`). 
# Additionally, select rows in which `gh_tenure` is greater than or equal to 1,991. 
(gh_turnover
 .filter(regex = "e$|s")  # select desired columns
 .query("gh_tenure >= 1991") # keep rows based on value the criteria     
)
```

```python
# Use the query you just performed. You will calculate the mean and median of `user_commits`. 
(gh_turnover
 .filter(regex = "e$|s")  # select desired columns
 .query("gh_tenure >= 1991") # keep rows based on value the criteria    
 .filter(["user_commits"])
 .agg(["mean","median"])
)
```
