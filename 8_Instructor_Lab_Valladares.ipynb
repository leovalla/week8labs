## Sorting and Ordering Data

```python
# Select the columns `Team_type`, `human_work`, and `work_per_human`.
(gh_teams
 .filter(["Team_type", "human_work", "work_per_human"])  # select desired columns
 .head()  # display first n rows
)
```

```python
# Select columns that end in the letter `t`. Use the regex `t$`.
(gh_teams
 .filter(regex = "t$")  # select desired columns
 .head(6)  # display first n rows
)
```

```python
# Sort your data descending using the columns `Team_size_class`, `human_work`, `work_per_human`.
(gh_teams
 .sort_values(by = ["Team_type", "human_work", "work_per_human"], ascending = False) # sort the data largest to smallest 
 .head(10)  # display first n rows
)
```

```python
# Select `human-bot` teams that have a `bot_members_count` value greater than and equal to 3.
(gh_teams
 .query("Team_type == 'human-bot'  & bot_members_count >= 3") # keep rows based on value the criteria 
 .head(6)  # display first n rows
)
```

```python
# Find the `human` teams that are `Large` and have a `human_gini` value greater than and equal to 0.75.
(gh_teams
 .query("Team_type == 'human' & Team_size_class == 'Large' & human_gini >= 0.75")  # keep rows based on value the criteria 
 .head(6)  # display first n rows
)
```

```python
# How many teams are in the `Small` or `Large` category?
(gh_teams
 .query("Team_size_class == 'Large' |Team_size_class == 'Small'")  # keep rows based on value in specified range
 .agg(["count"])
)
```

There are 428 `Small` or `Large` teams 

```python
# How many teams are in the `Small` or `Large` category with a `human_gini` value less than and equal to 0.25?

(gh_teams
 .query("(Team_size_class == 'Large' | Team_size_class == 'Small') & human_gini <= 0.25 ")  # keep rows based on value in specified range
 .agg(["count"])
)
```

There are 89 `Small` or `Large` teams with a `human_gini` value less than and equal to 0.25

```python
# How many `human` teams are in the `Medium` category?

(gh_teams
 .query("Team_type == 'human' & Team_size_class == 'Medium'")  # keep rows based on value in specified range
 .agg(["count"])
)
```

There are 96 `human` teams with a `Medium` category
